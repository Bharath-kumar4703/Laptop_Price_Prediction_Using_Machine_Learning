# **DATA PREPROCESSING**

```python
# Removing the un useful columns
df=df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1)
```

```python
df.head(2)
```

```python
df.columns
```

```python
# Fill null values
df['Company'].fillna(df['Company'].mode()[0],inplace=True)
df['TypeName'].fillna(df['TypeName'].mode()[0],inplace=True)
df['Inches'].fillna(df['Inches'].mode()[0],inplace=True)
df['ScreenResolution'].fillna(df['ScreenResolution'].mode()[0],inplace=True)
df['Cpu'].fillna(df['Cpu'].mode()[0],inplace=True)
df['Ram'].fillna(df['Ram'].mode()[0],inplace=True)
df['Memory'].fillna(df['Memory'].mode()[0],inplace=True)
df['Gpu'].fillna(df['Gpu'].mode()[0],inplace=True)
df['OpSys'].fillna(df['OpSys'].mode()[0],inplace=True)
df['Weight'].fillna(df['Weight'].mode()[0],inplace=True)
df['Price'].fillna(df['Price'].mean(),inplace=True)
```

```python
# Categorical_cols

categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols
```

```python
# Numerical_cols

numerical_cols = df.select_dtypes(include=['int64','float64']).columns
numerical_cols
```

```python
# fill missing values

for cl in categorical_cols:
  df[cl].fillna(df[cl].mode()[0],inplace=True)

for nl in numerical_cols:
  df[nl].fillna(df[nl].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# Drop the duplicates

df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Converting object to float datatype

df['Inches'] = pd.to_numeric(df['Inches'], errors='coerce').astype('float')
```

```python
# Remove GB in Ram column
df['Ram'] = df['Ram'].astype('str')
df['Ram'] = df['Ram'].str.replace('GB','',regex=False)
df['Ram'] = pd.to_numeric(df['Ram'], errors='coerce').astype('int')

# Remove kg in Weight column
df['Weight'] = df['Weight'].astype('str')
df['Weight'] = df['Weight'].str.replace('kg','',regex=False)
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce').astype('float')
```

# **Feature Engineering**

```python
from sklearn.preprocessing import OneHotEncoder

# 1. Select categorical columns

categorical_columns = ['Company', 'TypeName', 'Cpu', 'Gpu', 'OpSys']

# 3. Encode the categorical columns using OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop="first", handle_unknown='ignore')#handle_unknown='ignore will ignore any unknown categories during the trans
encoded_data = encoder.fit_transform(df[categorical_columns])

# 4. Create a DataFrame with encoded column names
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_columns))

# 5. Reset index before combining DataFrames
df = df.reset_index(drop=True)
encoded_df = encoded_df.reset_index(drop=True)

# 6. Drop original categorical columns and concatenate with encoded ones
df = df.drop(columns=['Company', 'TypeName', 'Cpu', 'Gpu', 'OpSys'])
final_data = pd.concat([df, encoded_df], axis=1)

final_data.head()
```

```python
final_data.select_dtypes(exclude=['object']).columns
```

```python
# Then We can do the Feature Scaling for better performance of machine learning algorithm
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
final_scaling_data = ss.fit_transform(final_data)
final_scaling_data=pd.DataFrame(data=final_scaling_data, columns=final_data.columns)
final_scaling_data.head()
```

# **Model Development**

```python
from sklearn.model_selection import train_test_split
x = final_scaling_data.drop(columns=['Price'])
y = final_scaling_data['Price']
#Train-Test Split
x_train, x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.impute import SimpleImputer

# Create the imputer
imputer = SimpleImputer(strategy='mean')  # or 'median' or 'most_frequent'

# Fit and transform training data
x_train = imputer.fit_transform(x_train)

# Transform test data with the same imputer
x_test = imputer.transform(x_test)
```

```python
# Train the models and test(predict) the output data
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Linear_Regression_Model
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
# Linear_Regression_Model Prediction
y_pred = lr_model.predict(x_test)
# Find out the accuracy
lr_rmse = np.sqrt(mean_squared_error(y_test,y_pred))
lr_r2 =r2_score(y_test, y_pred)


# RandomForestRegressor_Model
rf_model = RandomForestRegressor (random_state=42)
rf_model.fit(x_train, y_train)
# RandomForestRegressor_Model Prediction
rf_pred = rf_model.predict(x_test)
# Find out the accuracy
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)


# GradientBoostingRegressor_Model
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(x_train,y_train)
# GradientBoostingRegressor_Model Prediction
gb_pred = gb_model.predict(x_test)
# Find out the accuracy
gb_rmse= np.sqrt(mean_squared_error(y_test, gb_pred))
gb_r2 = r2_score(y_test, gb_pred)

```

```python
# Print the results
print(f"Linear Regression: RMSE = {lr_rmse:.2f}, R2 = {lr_r2:.2f}")
print(f"Random Forest: RMSE = {rf_rmse:.2f}, R2 = {rf_r2:.2f}")
print(f"Gradient Boosting: RMSE = {gb_rmse:.2f}, R2 = {gb_r2:.2f}")
```
